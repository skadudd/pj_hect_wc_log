```python
import pandas as pd
import json
from typing import List, Dict
import re
from urllib.parse import urlparse, parse_qs, unquote
from datetime import datetime

class LogToDetailedConverter:
    """JSON ë¡œê·¸ë¥¼ ë‹¤ì–‘í•œ Excel í˜•íƒœë¡œ ë³€í™˜í•˜ëŠ” í´ë˜ìŠ¤"""
    def __init__(self):
        self.input_data = []
        self.detailed_data = []
        self.detailed_deduplicated_data = []
        self.combined_data = []
        self.combined_data_no_comma = []  # ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„°
        self.combined_data_no_comma_deduplicated = []  # ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„° ì¤‘ë³µ ì œê±°
        self.deduplicated_data = []

    def convert_log_to_all_formats(self, input_log_path: str, 
                                 detailed_output_path: str, 
                                 detailed_deduplicated_output_path: str,
                                 combined_output_path: str,
                                 combined_deduplicated_output_path: str):
        """
        JSON ë¡œê·¸ íŒŒì¼ì„ ìƒì„¸ í˜•íƒœì™€ ê²°í•© í˜•íƒœ ë‘ ê°€ì§€ë¡œ ë³€í™˜
        
        Args:
            input_log_path: log_file.txt íŒŒì¼ ê²½ë¡œ
            detailed_output_path: log_analysis_detailed.xlsx ì¶œë ¥ íŒŒì¼ ê²½ë¡œ (ê° key-valueê°€ ê°œë³„ í–‰)
            combined_output_path: log_analysis_combined.xlsx ì¶œë ¥ íŒŒì¼ ê²½ë¡œ (key, valueê°€ ì‰¼í‘œë¡œ êµ¬ë¶„ëœ ì—´)
        """
        try:
            # 1. ì…ë ¥ ë¡œê·¸ íŒŒì¼ ì½ê¸°
            self._load_log_file(input_log_path)
            
            # 2. ìƒì„¸ ë°ì´í„°ë¡œ ë³€í™˜
            self._convert_to_detailed()
            
            # 3. ê²°í•© ë°ì´í„°ë¡œ ë³€í™˜
            self._convert_to_combined()
            
            # 4. ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±°
            self._convert_to_combined_deduplicated()
            
            # 5. ìƒì„¸ í˜•íƒœ ì¤‘ë³µ ì œê±°
            self._convert_to_detailed_deduplicated()
            
            # 6. ë„¤ í˜•íƒœ ëª¨ë‘ ì €ì¥
            self._save_detailed_excel(detailed_output_path)
            self._save_detailed_deduplicated_excel(detailed_deduplicated_output_path)
            self._save_combined_excel(combined_output_path)
            self._save_combined_deduplicated_excel(combined_deduplicated_output_path)
            
        except Exception as e:
            print(f"âŒ ë³€í™˜ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")

    def convert_log_to_combined_no_comma(self, input_log_path: str, 
                                       detailed_output_path: str,
                                       combined_no_comma_output_path: str,
                                       combined_no_comma_deduplicated_output_path: str = None):
        """
        JSON ë¡œê·¸ íŒŒì¼ì„ ìƒì„¸ í˜•íƒœì™€ ì‰¼í‘œ ì œê±°ëœ ê²°í•© í˜•íƒœë¡œ ë³€í™˜ (deduplicated ë²„ì „ í¬í•¨)
        
        Args:
            input_log_path: log_file.txt íŒŒì¼ ê²½ë¡œ
            detailed_output_path: log_analysis_detailed.xlsx ì¶œë ¥ íŒŒì¼ ê²½ë¡œ
            combined_no_comma_output_path: ì‰¼í‘œ ì œê±°ëœ ê²°í•© í˜•íƒœ ì¶œë ¥ íŒŒì¼ ê²½ë¡œ
            combined_no_comma_deduplicated_output_path: ì‰¼í‘œ ì œê±°ëœ ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±° ì¶œë ¥ íŒŒì¼ ê²½ë¡œ (ì„ íƒ)
        """
        try:
            # 1. ì…ë ¥ ë¡œê·¸ íŒŒì¼ ì½ê¸°
            self._load_log_file(input_log_path)
            
            # 2. ìƒì„¸ ë°ì´í„°ë¡œ ë³€í™˜
            self._convert_to_detailed()
            
            # 3. ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„°ë¡œ ë³€í™˜
            self._convert_to_combined_no_comma()
            
            # 4. ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„° ì¤‘ë³µ ì œê±°
            if combined_no_comma_deduplicated_output_path:
                self._convert_to_combined_no_comma_deduplicated()
            
            # 5. íŒŒì¼ ì €ì¥
            self._save_detailed_excel(detailed_output_path)
            self._save_combined_no_comma_excel(combined_no_comma_output_path)
            
            if combined_no_comma_deduplicated_output_path:
                self._save_combined_no_comma_deduplicated_excel(combined_no_comma_deduplicated_output_path)
            
            print(f"âœ… ì‰¼í‘œ ì œê±°ëœ ê²°í•© í˜•íƒœ ë³€í™˜ ì™„ë£Œ")
            
        except Exception as e:
            print(f"âŒ ë³€í™˜ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")

    def convert_log_to_both_formats(self, input_log_path: str, 
                                  detailed_output_path: str, 
                                  combined_output_path: str):
        """
        ê¸°ì¡´ í˜¸í™˜ì„±ì„ ìœ„í•œ ë©”ì„œë“œ - ì¤‘ë³µ ì œê±° ì—†ì´ ë‘ ê°€ì§€ í˜•íƒœë§Œ ë³€í™˜
        
        Args:
            input_log_path: log_file.txt íŒŒì¼ ê²½ë¡œ
            detailed_output_path: log_analysis_detailed.xlsx ì¶œë ¥ íŒŒì¼ ê²½ë¡œ (ê° key-valueê°€ ê°œë³„ í–‰, log_time í¬í•¨)
            combined_output_path: log_analysis_combined.xlsx ì¶œë ¥ íŒŒì¼ ê²½ë¡œ (key, valueê°€ ì‰¼í‘œë¡œ êµ¬ë¶„ëœ ì—´, log_time í¬í•¨)
        """
        try:
            # 1. ì…ë ¥ ë¡œê·¸ íŒŒì¼ ì½ê¸°
            self._load_log_file(input_log_path)
            
            # 2. ìƒì„¸ ë°ì´í„°ë¡œ ë³€í™˜
            self._convert_to_detailed()
            
            # 3. ê²°í•© ë°ì´í„°ë¡œ ë³€í™˜
            self._convert_to_combined()
            
            # 4. ë‘ í˜•íƒœ ì €ì¥
            self._save_detailed_excel(detailed_output_path)
            self._save_combined_excel(combined_output_path)
            
        except Exception as e:
            print(f"âŒ ë³€í™˜ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")

    def _load_log_file(self, log_path: str):
        """JSON ë¡œê·¸ íŒŒì¼ ë¡œë“œ ë° íŒŒì‹± (ì˜¤ë¥˜ ìë™ ìˆ˜ì • ê¸°ëŠ¥ í¬í•¨)"""
        try:
            self.input_data = []
            json_errors = []
            fixed_count = 0
            
            with open(log_path, 'r', encoding='utf-8') as file:
                for line_no, line in enumerate(file, 1):
                    line = line.strip()
                    if not line:
                        continue
                    
                    try:
                        # JSON íŒŒì‹± (ìƒˆë¡œìš´ í˜•ì‹: ì§ì ‘ key-value ê°ì²´)
                        log_entry = json.loads(line)
                        
                        # ìƒˆë¡œìš´ í˜•ì‹ ì²˜ë¦¬: ì§ì ‘ key-value ê°ì²´ë¡œ ì œê³µë¨
                        parsed_data = self._process_new_log_format(log_entry)
                        
                        if parsed_data:
                            # ê¸°ë³¸ ë¡œê·¸ ì •ë³´ ì¶”ê°€
                            parsed_data['log_line'] = line_no
                            
                            self.input_data.append(parsed_data)
                    
                    except json.JSONDecodeError as e:
                        json_errors.append((line_no, str(e), line))
                        
                        # JSON ìˆ˜ì • ì‹œë„
                        fixed_line = self._try_fix_json(line)
                        if fixed_line:
                            try:
                                log_entry = json.loads(fixed_line)
                                
                                # ìƒˆë¡œìš´ í˜•ì‹ ì²˜ë¦¬
                                parsed_data = self._process_new_log_format(log_entry)
                                
                                if parsed_data:
                                    # ê¸°ë³¸ ë¡œê·¸ ì •ë³´ ì¶”ê°€
                                    parsed_data['log_line'] = line_no
                                    
                                    self.input_data.append(parsed_data)
                                    fixed_count += 1
                                    print(f"âœ… JSON ìˆ˜ì • ì„±ê³µ (ë¼ì¸ {line_no})")
                            except:
                                # ì¶”ê°€ ìˆ˜ì • ì‹œë„
                                fixed_line2 = self._force_fix_json(line)
                                if fixed_line2:
                                    try:
                                        log_entry = json.loads(fixed_line2)
                                        
                                        # ìƒˆë¡œìš´ í˜•ì‹ ì²˜ë¦¬
                                        parsed_data = self._process_new_log_format(log_entry)
                                        
                                        if parsed_data:
                                            # ê¸°ë³¸ ë¡œê·¸ ì •ë³´ ì¶”ê°€
                                            parsed_data['log_line'] = line_no
                                            
                                            self.input_data.append(parsed_data)
                                            fixed_count += 1
                                            print(f"âœ… ê°•ì œ JSON ìˆ˜ì • ì„±ê³µ (ë¼ì¸ {line_no})")
                                    except:
                                        print(f"âŒ JSON ìˆ˜ì • ì‹¤íŒ¨ (ë¼ì¸ {line_no}): {e}")
                                else:
                                    print(f"âŒ JSON ìˆ˜ì • ì‹¤íŒ¨ (ë¼ì¸ {line_no}): {e}")
                        else:
                            print(f"âŒ JSON ìˆ˜ì • ì‹¤íŒ¨ (ë¼ì¸ {line_no}): {e}")
            
            # ì˜¤ë¥˜ ìš”ì•½ ì¶œë ¥
            if json_errors:
                print(f"âš ï¸ JSON íŒŒì‹± ì˜¤ë¥˜ ë°œìƒ: {len(json_errors)}ê°œ")
                if fixed_count > 0:
                    print(f"âœ… ìˆ˜ì • ì„±ê³µ: {fixed_count}ê°œ")
                print(f"âœ… ìµœì¢… íŒŒì‹±ëœ ë¡œê·¸: {len(self.input_data)}ê°œ")
            else:
                print(f"âœ… ëª¨ë“  JSON íŒŒì‹± ì„±ê³µ")
            
            print(f"âœ… ë¡œê·¸ íŒŒì¼ ë¡œë“œ ì™„ë£Œ: {len(self.input_data)}ê°œ ë¡œê·¸")
            
        except FileNotFoundError:
            print(f"âŒ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {log_path}")
            raise
        except Exception as e:
            print(f"âŒ íŒŒì¼ ë¡œë“œ ì˜¤ë¥˜: {e}")
            raise

    def _process_new_log_format(self, log_entry: dict) -> dict:
        """ìƒˆë¡œìš´ ë¡œê·¸ í˜•ì‹ ì²˜ë¦¬: ì§ì ‘ key-value ê°ì²´, ë¹ˆ ê°’ ì œì™¸, íŠ¹ì • í•„ë“œ ì œì™¸"""
        # ì œì™¸í•  í•„ë“œë“¤ (ì‚¬ìš©ì ìš”ì²­)
        exclude_fields = {'client_ip', 'os_version', 'platform_type', 'device_model', 'device_maker', 'event_dt', 'nth_pcid', 'nth_sid', 'user_seq'}
        
        # ì²˜ë¦¬ëœ ë°ì´í„°
        processed_data = {}
        
        # event_dttmì„ logtimeìœ¼ë¡œ ë³€í™˜ (ê¸°ì¡´ ì½”ë“œ í˜¸í™˜ì„±ì„ ìœ„í•´)
        if 'event_dttm' in log_entry:
            processed_data['logtime'] = log_entry['event_dttm']
        
        # ê° key-value ìŒ ì²˜ë¦¬
        for key, value in log_entry.items():
            # ì œì™¸ í•„ë“œëŠ” ê±´ë„ˆë›°ê¸°
            if key in exclude_fields:
                continue
            
            # event_dttmì€ ì´ë¯¸ logtimeìœ¼ë¡œ ì²˜ë¦¬í–ˆìœ¼ë¯€ë¡œ ê±´ë„ˆë›°ê¸°
            if key == 'event_dttm':
                continue
            
            # ë¹ˆ ë¬¸ìì—´("")ì¸ ê°’ì€ ì œì™¸
            if value == "":
                continue
            
            # ìœ íš¨í•œ key-value ìŒë§Œ í¬í•¨
            processed_data[key] = value
        
        return processed_data

    def _try_fix_json(self, line: str) -> str:
        """JSON ìˆ˜ì • ì‹œë„ - 1ì°¨ ìˆ˜ì •"""
        try:
            # 1. ë§ˆì§€ë§‰ ì‰¼í‘œ ì œê±°
            if line.endswith(',}'):
                line = line[:-2] + '}'
            elif line.endswith(',]'):
                line = line[:-2] + ']'
            
            # 2. ê°’ ë‚´ë¶€ì˜ ë”°ì˜´í‘œë¥¼ ì‘ì€ë”°ì˜´í‘œë¡œ ë³€ê²½
            import re
            
            # ëª¨ë“  "key":"value with quotes" íŒ¨í„´ì„ ì°¾ì•„ì„œ ìˆ˜ì •
            def fix_quotes_in_value(match):
                key = match.group(1)
                value = match.group(2)
                # ê°’ ë‚´ë¶€ì˜ ëª¨ë“  ë”°ì˜´í‘œë¥¼ ì‘ì€ë”°ì˜´í‘œë¡œ ë³€ê²½
                fixed_value = value.replace('"', "'")
                return f'"{key}":"{fixed_value}"'
            
            # ê°’ì— ë”°ì˜´í‘œê°€ í¬í•¨ëœ íŒ¨í„´ ì°¾ê¸°
            problem_pattern = r'"([^"]+)":"([^"]*"[^"]*)"'
            
            # ì—¬ëŸ¬ ë²ˆ ì ìš©í•˜ì—¬ ëª¨ë“  ë¬¸ì œ í•´ê²°
            prev_line = ""
            iterations = 0
            while line != prev_line and iterations < 5:
                prev_line = line
                line = re.sub(problem_pattern, fix_quotes_in_value, line)
                iterations += 1
            
            return line
            
        except Exception:
            return None

    def _force_fix_json(self, line: str) -> str:
        """JSON ê°•ì œ ìˆ˜ì • - 2ì°¨ ìˆ˜ì •"""
        try:
            # ê¸°ë³¸ ì •ë¦¬
            if line.endswith(',}'):
                line = line[:-2] + '}'
            elif line.endswith(',]'):
                line = line[:-2] + ']'
            
            # JSON êµ¬ì¡° ë¶„ì„í•˜ì—¬ í‚¤-ê°’ ìŒë³„ë¡œ ì²˜ë¦¬
            if not (line.startswith('{') and line.endswith('}')):
                return None
            
            # ì¤‘ê´„í˜¸ ì œê±°
            content = line[1:-1]
            
            # í‚¤-ê°’ ìŒë“¤ì„ ë¶„ë¦¬
            pairs = []
            current_pair = ""
            bracket_count = 0
            in_quotes = False
            escape_next = False
            
            for char in content:
                if escape_next:
                    current_pair += char
                    escape_next = False
                    continue
                
                if char == '\\':
                    current_pair += char
                    escape_next = True
                    continue
                
                if char == '"' and not escape_next:
                    in_quotes = not in_quotes
                
                if not in_quotes:
                    if char in '{}[]':
                        bracket_count += 1 if char in '{[' else -1
                    elif char == ',' and bracket_count == 0:
                        pairs.append(current_pair.strip())
                        current_pair = ""
                        continue
                
                current_pair += char
            
            if current_pair.strip():
                pairs.append(current_pair.strip())
            
            # ê° í‚¤-ê°’ ìŒì„ ìˆ˜ì •
            fixed_pairs = []
            for pair in pairs:
                if ':' in pair:
                    try:
                        # í‚¤ì™€ ê°’ ë¶„ë¦¬ (ì²« ë²ˆì§¸ ì½œë¡  ê¸°ì¤€)
                        colon_pos = pair.find(':')
                        key_part = pair[:colon_pos].strip()
                        value_part = pair[colon_pos+1:].strip()
                        
                        # ê°’ ë¶€ë¶„ì—ì„œ ë”°ì˜´í‘œë¡œ ë‘˜ëŸ¬ì‹¸ì¸ ë¬¸ìì—´ì¸ì§€ í™•ì¸
                        if value_part.startswith('"') and value_part.endswith('"'):
                            # ê°’ ë‚´ë¶€ì˜ ë”°ì˜´í‘œë¥¼ ì‘ì€ë”°ì˜´í‘œë¡œ ë³€ê²½
                            inner_value = value_part[1:-1]  # ì–‘ìª½ ë”°ì˜´í‘œ ì œê±°
                            fixed_inner = inner_value.replace('"', "'")
                            fixed_value = f'"{fixed_inner}"'
                            fixed_pairs.append(f'{key_part}:{fixed_value}')
                        else:
                            fixed_pairs.append(pair)
                    except:
                        fixed_pairs.append(pair)
                else:
                    fixed_pairs.append(pair)
            
            # ë‹¤ì‹œ ì¡°í•©
            result = '{' + ','.join(fixed_pairs) + '}'
            return result
            
        except Exception:
            return None

    def _parse_request_url(self, request_url: str) -> Dict:
        """REQUEST URLì—ì„œ íŒŒë¼ë¯¸í„°ë“¤ì„ ì¶”ì¶œ (ê°œì„ ëœ í•œê¸€ ì²˜ë¦¬)"""
        if not request_url:
            return {}
        
        try:
            # ìœ ë‹ˆì½”ë“œ ì´ìŠ¤ì¼€ì´í”„ ë””ì½”ë”©
            url = request_url.encode('utf-8').decode('unicode_escape')
            parsed_url = urlparse(url)
            
            # ì¿¼ë¦¬ íŒŒë¼ë¯¸í„°ë¥¼ ìˆ˜ë™ìœ¼ë¡œ íŒŒì‹± (ë³µì¡í•œ ê°’ ì²˜ë¦¬)
            query = parsed_url.query
            processed_params = {}
            
            if not query:
                return processed_params
            
            # &ë¡œ ë¶„í• í•˜ë˜, URL ì¸ì½”ë”©ëœ &(%26)ì€ ë¶„í• í•˜ì§€ ì•ŠìŒ
            params = []
            current_param = ""
            i = 0
            
            while i < len(query):
                if query[i:i+1] == '&' and not query[max(0, i-3):i] == '%26':
                    if current_param:
                        params.append(current_param)
                    current_param = ""
                else:
                    current_param += query[i]
                i += 1
            
            if current_param:
                params.append(current_param)
            
            # ê° íŒŒë¼ë¯¸í„°ë¥¼ í‚¤=ê°’ìœ¼ë¡œ ë¶„í• 
            for param in params:
                if '=' not in param:
                    continue
                    
                # ì²« ë²ˆì§¸ = ê¸°ì¤€ìœ¼ë¡œë§Œ ë¶„í•  (ê°’ì— =ì´ í¬í•¨ë  ìˆ˜ ìˆìŒ)
                key, value = param.split('=', 1)
                
                # í‚¤ì™€ ê°’ ë””ì½”ë”©
                decoded_key = self._fix_korean_text(key)
                decoded_value = self._fix_korean_text(value)
                
                processed_params[decoded_key] = decoded_value
            
            return processed_params
            
        except Exception as e:
            print(f"URL íŒŒì‹± ì˜¤ë¥˜: {e}")
            # ì‹¤íŒ¨ ì‹œ ê¸°ì¡´ ë°©ì‹ìœ¼ë¡œ fallback
            try:
                url = request_url.encode('utf-8').decode('unicode_escape')
                parsed_url = urlparse(url)
                query_params = parse_qs(parsed_url.query, keep_blank_values=True)
                
                processed_params = {}
                for key, value_list in query_params.items():
                    raw_value = value_list[0] if value_list else ''
                    decoded_value = self._fix_korean_text(raw_value)
                    processed_params[key] = decoded_value
                
                return processed_params
            except:
                return {}

    def _fix_korean_text(self, text: str) -> str:
        """ê¹¨ì§„ í•œê¸€ì„ ì˜¬ë°”ë¥´ê²Œ ë””ì½”ë”©"""
        try:
            # URL ë””ì½”ë”© ë¨¼ì € ì‹œë„
            from urllib.parse import unquote
            decoded = unquote(text, encoding='utf-8')
            
            # ê¹¨ì§„ í•œê¸€ íŒ¨í„´ì´ ìˆëŠ”ì§€ í™•ì¸
            if self._has_broken_korean(decoded):
                # Latin-1ë¡œ ì¸ì½”ë”© í›„ UTF-8ë¡œ ë””ì½”ë”© (ì¼ë°˜ì ì¸ mojibake í•´ê²°ë²•)
                try:
                    fixed = decoded.encode('latin-1').decode('utf-8')
                    return fixed
                except:
                    pass
                    
                # ê·¸ë˜ë„ ì•ˆë˜ë©´ ë°”ì´íŠ¸ ë‹¨ìœ„ë¡œ ì²˜ë¦¬
                try:
                    # ë¬¸ìì—´ì˜ ê° ë¬¸ìë¥¼ ë°”ì´íŠ¸ë¡œ ë³€í™˜ í›„ UTF-8 ë””ì½”ë”©
                    byte_data = bytes([ord(c) for c in decoded if ord(c) < 256])
                    fixed = byte_data.decode('utf-8', errors='ignore')
                    return fixed
                except:
                    pass
                    
                # UTF-8 ë°”ì´íŠ¸ íŒ¨í„´ìœ¼ë¡œ ë³µêµ¬ ì‹œë„
                try:
                    # íŠ¹ì • ê¹¨ì§„ íŒ¨í„´ë“¤ì„ ì§ì ‘ ë§¤í•‘
                    broken_patterns = {
                        'ÃªÂ·Â¹Ã¬Ã¬Ã«ÂªÂ¨': 'ê·¹ì¥ìš©ëª¨',
                        'Ã­Â¥ÃªÂ¸Â°Ã«Â¡Â­ÃªÂ²': 'í–¥ê¸°ë¡­ê²Œ',
                        'Ã¬Â¤ÃªÂ¸Â°Ã¬Â¼Ã¬Â´': 'ìˆ˜ê¸°ì¼ì´',
                        'Ã¬Ã¬Ã«ÂªÂ¨': 'ìš©ëª¨',
                        'ÃªÂ·Â¹': 'ê·¹'
                    }
                    
                    result = decoded
                    for broken, fixed in broken_patterns.items():
                        result = result.replace(broken, fixed)
                    
                    if result != decoded:
                        return result
                        
                except:
                    pass
            
            return decoded
            
        except Exception:
            return text

    def _has_broken_korean(self, text: str) -> bool:
        """ê¹¨ì§„ í•œê¸€ íŒ¨í„´ì´ ìˆëŠ”ì§€ í™•ì¸"""
        broken_patterns = [
            'Ã«Â¶', 'Ã¬Ã«', 'ÃªÂ¼Â¬', 'Ã«', 'Ã­Â¼', 'Ã¬Â¸', 'ÃªÂ¸Â°', 'ÃªÂ¸', 'Ã¬Ã¬Â¹', 'Ã¬', 'Ãª',
            'ÃªÂ·Â¹Ã¬', 'Ã­Â¥ÃªÂ¸Â°', 'Ã¬Â¤ÃªÂ¸Â°', 'Ã¬Ã¬Ã«', 'Ã«ÂªÂ¨'
        ]
        return any(pattern in text for pattern in broken_patterns)

    def _format_logtime(self, logtime_str: str) -> str:
        """LOGTIMEì„ yyyy-mm-dd HH:MM:SS í˜•ì‹ìœ¼ë¡œ ë³€í™˜ (ê¸°ì¡´ í˜•ì‹ê³¼ ìƒˆ í˜•ì‹ ëª¨ë‘ ì§€ì›)"""
        try:
            if not logtime_str:
                return ''
            
            # ìƒˆë¡œìš´ í˜•ì‹: "2025-06-05T15:04:00.041" (ISO 8601)
            if 'T' in logtime_str:
                try:
                    # ISO 8601 í˜•ì‹ íŒŒì‹±
                    if '.' in logtime_str:
                        # ë§ˆì´í¬ë¡œì´ˆ í¬í•¨ëœ ê²½ìš°
                        dt = datetime.fromisoformat(logtime_str.replace('Z', '+00:00'))
                    else:
                        dt = datetime.fromisoformat(logtime_str.replace('Z', '+00:00'))
                    
                    # yyyy-mm-dd HH:MM:SS í˜•ì‹ìœ¼ë¡œ ë³€í™˜
                    formatted_time = dt.strftime('%Y-%m-%d %H:%M:%S')
                    return formatted_time
                except:
                    # ISO í˜•ì‹ íŒŒì‹± ì‹¤íŒ¨ ì‹œ ê¸°ì¡´ ë°©ì‹ìœ¼ë¡œ fallback
                    pass
            
            # ê¸°ì¡´ í˜•ì‹: "04/Jun/2025:21:46:25 +0900"
            # ì›” ì´ë¦„ì„ ìˆ«ìë¡œ ë§¤í•‘
            month_map = {
                'Jan': '01', 'Feb': '02', 'Mar': '03', 'Apr': '04',
                'May': '05', 'Jun': '06', 'Jul': '07', 'Aug': '08',
                'Sep': '09', 'Oct': '10', 'Nov': '11', 'Dec': '12'
            }
            
            # ì •ê·œì‹ìœ¼ë¡œ íŒŒì‹±: "04/Jun/2025:21:46:25 +0900"
            pattern = r'(\d{2})/(\w{3})/(\d{4}):(\d{2}):(\d{2}):(\d{2})\s*\+\d{4}'
            match = re.match(pattern, logtime_str)
            
            if match:
                day, month_name, year, hour, minute, second = match.groups()
                month = month_map.get(month_name, '01')
                
                # yyyy-mm-dd HH:MM:SS í˜•ì‹ìœ¼ë¡œ ë³€í™˜
                formatted_time = f"{year}-{month}-{day} {hour}:{minute}:{second}"
                return formatted_time
            else:
                # íŒŒì‹± ì‹¤íŒ¨ ì‹œ ì›ë³¸ ë°˜í™˜
                return logtime_str
                
        except Exception as e:
            print(f"âš ï¸ LOGTIME ë³€í™˜ ì˜¤ë¥˜: {e} - ì›ë³¸: {logtime_str}")
            return logtime_str

    def _convert_to_detailed(self):
        """ë¡œê·¸ ë°ì´í„°ë¥¼ ìƒì„¸ key-value í–‰ìœ¼ë¡œ ë³€í™˜"""
        self.detailed_data = []
        
        # ì œì™¸í•  ê¸°ë³¸ í•„ë“œë“¤ (ë©”íƒ€ë°ì´í„°)
        exclude_fields = {'log_line', 'logtime'}
        
        for log_entry in self.input_data:
            log_no = log_entry.get('log_line', 0)
            
            # ê¸°ë³¸ ì •ë³´ ì¶”ì¶œ
            page_id = log_entry.get('page_id', '')
            click_type = log_entry.get('click_type', '')
            act_type = log_entry.get('act_type', '')
            click_text = log_entry.get('click_text', '')
            
            # LOGTIME í¬ë§· ë³€í™˜
            raw_logtime = log_entry.get('logtime', '')
            formatted_logtime = self._format_logtime(raw_logtime)
            
            # ë‚˜ë¨¸ì§€ íŒŒë¼ë¯¸í„°ë“¤ì„ key-value ìŒìœ¼ë¡œ ì²˜ë¦¬
            for key, value in log_entry.items():
                if key in exclude_fields:
                    continue
                
                # ê¸°ë³¸ í•„ë“œë“¤ë„ key-value ìŒì— í¬í•¨ (ì‚¬ìš©ì ìš”ì²­)
                # page_id, click_type, act_typeë„ ì´ì œ key-valueë¡œ í‘œì‹œë¨
                
                self.detailed_data.append({
                    'no': log_no,
                    'log_time': formatted_logtime,
                    'page_id': page_id,
                    'click_type': click_type,
                    'act_type': act_type,
                    'click_text': click_text,
                    'key': key,
                    'value': str(value) if value is not None else ''
                })
        
        print(f"âœ… ìƒì„¸ ë°ì´í„° ë³€í™˜ ì™„ë£Œ: {len(self.detailed_data)}ê°œ key-value ìŒ")

    def fill_missing_rows_in_combined_data(self):
        """combined_dataì—ì„œ ëˆ„ë½ëœ no ë²ˆí˜¸ì— ë¹ˆ í–‰ì„ ì¶”ê°€"""
        if not self.combined_data:
            print("âš ï¸ ê²°í•© ë°ì´í„°ê°€ ì—†ì–´ ëˆ„ë½ëœ í–‰ì„ ì±„ìš¸ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        # í˜„ì¬ no ê°’ë“¤ ì¶”ì¶œ
        current_nos = [item['no'] for item in self.combined_data]
        min_no = min(current_nos)
        max_no = max(current_nos)
        
        # ì „ì²´ ë²”ìœ„ì˜ no ë¦¬ìŠ¤íŠ¸ ìƒì„±
        full_range = set(range(min_no, max_no + 1))
        existing_nos = set(current_nos)
        missing_nos = sorted(full_range - existing_nos)
        
        if not missing_nos:
            print("âœ… ëˆ„ë½ëœ ë²ˆí˜¸ê°€ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        print(f"ğŸ“Œ ëˆ„ë½ëœ ë²ˆí˜¸: {missing_nos}")
        
        # ê¸°ë³¸ ë¹ˆ í–‰ í…œí”Œë¦¿ ìƒì„± (ì²« ë²ˆì§¸ í–‰ì„ ê¸°ì¤€ìœ¼ë¡œ)
        template_row = self.combined_data[0].copy()
        
        # ëˆ„ë½ëœ ë²ˆí˜¸ë“¤ì— ëŒ€í•œ ë¹ˆ í–‰ ìƒì„±
        for missing_no in missing_nos:
            empty_row = template_row.copy()
            empty_row['no'] = missing_no
            empty_row['page_id'] = ''
            empty_row['click_type'] = ''
            empty_row['act_type'] = ''
            empty_row['keys_combined'] = ''
            empty_row['values_combined'] = ''
            empty_row['key_count'] = 0
            
            # click_textê°€ ìˆìœ¼ë©´ ë¹ˆ ê°’ìœ¼ë¡œ ì„¤ì •
            if 'click_text' in empty_row:
                empty_row['click_text'] = ''
            
            self.combined_data.append(empty_row)
        
        # no ìˆœìœ¼ë¡œ ì •ë ¬
        self.combined_data.sort(key=lambda x: x['no'])
        
        print(f"âœ… ëˆ„ë½ëœ {len(missing_nos)}ê°œ í–‰ì„ ì¶”ê°€í–ˆìŠµë‹ˆë‹¤.")
        print(f"   ì „ì²´ í–‰ ìˆ˜: {len(self.combined_data)}ê°œ (ì›ë³¸ + ì¶”ê°€ëœ ë¹ˆ í–‰)")

    def fill_missing_rows_in_combined_no_comma(self):
        """combined_data_no_commaì—ì„œ ëˆ„ë½ëœ no ë²ˆí˜¸ì— ë¹ˆ í–‰ì„ ì¶”ê°€"""
        if not self.combined_data_no_comma:
            print("âš ï¸ ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„°ê°€ ì—†ì–´ ëˆ„ë½ëœ í–‰ì„ ì±„ìš¸ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        # í˜„ì¬ no ê°’ë“¤ ì¶”ì¶œ
        current_nos = [item['no'] for item in self.combined_data_no_comma]
        min_no = min(current_nos)
        max_no = max(current_nos)
        
        # ì „ì²´ ë²”ìœ„ì˜ no ë¦¬ìŠ¤íŠ¸ ìƒì„±
        full_range = set(range(min_no, max_no + 1))
        existing_nos = set(current_nos)
        missing_nos = sorted(full_range - existing_nos)
        
        if not missing_nos:
            print("âœ… ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„°ì—ì„œ ëˆ„ë½ëœ ë²ˆí˜¸ê°€ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        print(f"ğŸ“Œ ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„°ì—ì„œ ëˆ„ë½ëœ ë²ˆí˜¸: {missing_nos}")
        
        # ê¸°ë³¸ ë¹ˆ í–‰ í…œí”Œë¦¿ ìƒì„±
        template_row = self.combined_data_no_comma[0].copy()
        
        # ëˆ„ë½ëœ ë²ˆí˜¸ë“¤ì— ëŒ€í•œ ë¹ˆ í–‰ ìƒì„±
        for missing_no in missing_nos:
            empty_row = template_row.copy()
            empty_row['no'] = missing_no
            empty_row['page_id'] = ''
            empty_row['click_type'] = ''
            empty_row['act_type'] = ''
            empty_row['keys_combined'] = ''
            empty_row['values_combined'] = ''
            empty_row['key_count'] = 0
            
            # click_textê°€ ìˆìœ¼ë©´ ë¹ˆ ê°’ìœ¼ë¡œ ì„¤ì •
            if 'click_text' in empty_row:
                empty_row['click_text'] = ''
            
            self.combined_data_no_comma.append(empty_row)
        
        # no ìˆœìœ¼ë¡œ ì •ë ¬
        self.combined_data_no_comma.sort(key=lambda x: x['no'])
        
        print(f"âœ… ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„°ì—ì„œ ëˆ„ë½ëœ {len(missing_nos)}ê°œ í–‰ì„ ì¶”ê°€í–ˆìŠµë‹ˆë‹¤.")

    def fill_missing_rows_in_detailed_data(self):
        """detailed_dataì—ì„œ ëˆ„ë½ëœ no ë²ˆí˜¸ì— ë¹ˆ í–‰ì„ ì¶”ê°€"""
        if not self.detailed_data:
            print("âš ï¸ ìƒì„¸ ë°ì´í„°ê°€ ì—†ì–´ ëˆ„ë½ëœ í–‰ì„ ì±„ìš¸ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        # í˜„ì¬ no ê°’ë“¤ ì¶”ì¶œ (ì¤‘ë³µ ì œê±°)
        current_nos = list(set([item['no'] for item in self.detailed_data]))
        min_no = min(current_nos)
        max_no = max(current_nos)
        
        # ì „ì²´ ë²”ìœ„ì˜ no ë¦¬ìŠ¤íŠ¸ ìƒì„±
        full_range = set(range(min_no, max_no + 1))
        existing_nos = set(current_nos)
        missing_nos = sorted(full_range - existing_nos)
        
        if not missing_nos:
            print("âœ… ìƒì„¸ ë°ì´í„°ì—ì„œ ëˆ„ë½ëœ ë²ˆí˜¸ê°€ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        print(f"ğŸ“Œ ìƒì„¸ ë°ì´í„°ì—ì„œ ëˆ„ë½ëœ ë²ˆí˜¸: {missing_nos}")
        
        # ê¸°ë³¸ ë¹ˆ í–‰ í…œí”Œë¦¿ ìƒì„±
        template_row = self.detailed_data[0].copy()
        
        # ëˆ„ë½ëœ ë²ˆí˜¸ë“¤ì— ëŒ€í•œ ë¹ˆ í–‰ ìƒì„± (ê° ë²ˆí˜¸ë‹¹ 1ê°œì˜ ë¹ˆ key-value ìŒ)
        for missing_no in missing_nos:
            empty_row = template_row.copy()
            empty_row['no'] = missing_no
            empty_row['log_time'] = ''
            empty_row['page_id'] = ''
            empty_row['click_type'] = ''
            empty_row['act_type'] = ''
            empty_row['click_text'] = ''
            empty_row['key'] = 'NO_LOG'  # ë¡œê·¸ê°€ ì—†ìŒì„ í‘œì‹œ
            empty_row['value'] = 'Missing Log Entry'  # ëˆ„ë½ëœ ë¡œê·¸ í‘œì‹œ
            
            self.detailed_data.append(empty_row)
        
        # no ìˆœìœ¼ë¡œ ì •ë ¬
        self.detailed_data.sort(key=lambda x: x['no'])
        
        print(f"âœ… ìƒì„¸ ë°ì´í„°ì—ì„œ ëˆ„ë½ëœ {len(missing_nos)}ê°œ í–‰ì„ ì¶”ê°€í–ˆìŠµë‹ˆë‹¤.")

    def fill_all_missing_rows(self):
        """ëª¨ë“  ë°ì´í„° íƒ€ì…ì—ì„œ ëˆ„ë½ëœ í–‰ì„ ì±„ì›€"""
        print("\n=== ëˆ„ë½ëœ í–‰ ì±„ìš°ê¸° ì‹œì‘ ===")
        
        self.fill_missing_rows_in_combined_data()
        self.fill_missing_rows_in_combined_no_comma()
        self.fill_missing_rows_in_detailed_data()
        
        print("=== ëˆ„ë½ëœ í–‰ ì±„ìš°ê¸° ì™„ë£Œ ===\n")

    def preview_missing_analysis(self):
        """ëˆ„ë½ëœ ë²ˆí˜¸ ë¶„ì„ ë¯¸ë¦¬ë³´ê¸°"""
        print("\nğŸ“Š ëˆ„ë½ëœ ë²ˆí˜¸ ë¶„ì„:")
        
        # Combined ë°ì´í„° ë¶„ì„
        if self.combined_data:
            current_nos = [item['no'] for item in self.combined_data]
            min_no, max_no = min(current_nos), max(current_nos)
            full_range = set(range(min_no, max_no + 1))
            existing_nos = set(current_nos)
            missing_nos = sorted(full_range - existing_nos)
            
            print(f"ğŸ“‹ Combined ë°ì´í„°:")
            print(f"   ë²”ìœ„: {min_no} ~ {max_no}")
            print(f"   ê¸°ì¡´ í–‰ ìˆ˜: {len(current_nos)}")
            print(f"   ëˆ„ë½ëœ ë²ˆí˜¸: {missing_nos if missing_nos else 'ì—†ìŒ'}")
        
        # Combined No Comma ë°ì´í„° ë¶„ì„
        if self.combined_data_no_comma:
            current_nos = [item['no'] for item in self.combined_data_no_comma]
            min_no, max_no = min(current_nos), max(current_nos)
            full_range = set(range(min_no, max_no + 1))
            existing_nos = set(current_nos)
            missing_nos = sorted(full_range - existing_nos)
            
            print(f"ğŸ“‹ Combined No Comma ë°ì´í„°:")
            print(f"   ë²”ìœ„: {min_no} ~ {max_no}")
            print(f"   ê¸°ì¡´ í–‰ ìˆ˜: {len(current_nos)}")
            print(f"   ëˆ„ë½ëœ ë²ˆí˜¸: {missing_nos if missing_nos else 'ì—†ìŒ'}")
        
        # Detailed ë°ì´í„° ë¶„ì„
        if self.detailed_data:
            current_nos = list(set([item['no'] for item in self.detailed_data]))
            min_no, max_no = min(current_nos), max(current_nos)
            full_range = set(range(min_no, max_no + 1))
            existing_nos = set(current_nos)
            missing_nos = sorted(full_range - existing_nos)
            
            print(f"ğŸ“‹ Detailed ë°ì´í„°:")
            print(f"   ë²”ìœ„: {min_no} ~ {max_no}")
            print(f"   ê¸°ì¡´ ë¡œê·¸ ìˆ˜: {len(current_nos)}")
            print(f"   ëˆ„ë½ëœ ë²ˆí˜¸: {missing_nos if missing_nos else 'ì—†ìŒ'}")

    def _convert_to_combined(self):
        """ë¡œê·¸ ë°ì´í„°ë¥¼ ê²°í•© í˜•íƒœë¡œ ë³€í™˜ (key, valueê°€ ì‰¼í‘œë¡œ êµ¬ë¶„ëœ ì—´)"""
        self.combined_data = []
        
        # ì œì™¸í•  ê¸°ë³¸ í•„ë“œë“¤ (ë©”íƒ€ë°ì´í„°)
        exclude_fields = {'log_line', 'logtime', 'clientip', 'useragent', 'status', 
                         'page_id', 'click_type', 'act_type'}
        
        for log_entry in self.input_data:
            log_no = log_entry.get('log_line', 0)
            
            # ê¸°ë³¸ ì •ë³´ ì¶”ì¶œ
            page_id = log_entry.get('page_id', '')
            click_type = log_entry.get('click_type', '')
            act_type = log_entry.get('act_type', '')
            click_text = log_entry.get('click_text', '')  # click_text ì¶”ê°€
            
            # ë‚˜ë¨¸ì§€ íŒŒë¼ë¯¸í„°ë“¤ ìˆ˜ì§‘ (ì´ì œ click_textë„ í¬í•¨ë¨)
            keys = []
            values = []
            
            for key, value in log_entry.items():
                if key not in exclude_fields:
                    keys.append(key)
                    values.append(str(value) if value is not None else '')
            
            # í‚¤ì™€ ê°’ì„ ì‰¼í‘œë¡œ êµ¬ë¶„ëœ ë¬¸ìì—´ë¡œ ê²°í•©
            keys_combined = ', '.join(keys) if keys else ''
            values_combined = ', '.join(values) if values else ''
            
            combined_entry = {
                'no': log_no,
                'page_id': page_id,
                'click_type': click_type,
                'act_type': act_type,
                'keys_combined': keys_combined,
                'values_combined': values_combined,
                'key_count': len(keys)
            }
            
            # click_textê°€ ìˆìœ¼ë©´ ì¶”ê°€
            if click_text:
                combined_entry['click_text'] = click_text
            
            self.combined_data.append(combined_entry)
        
        print(f"âœ… ê²°í•© ë°ì´í„° ë³€í™˜ ì™„ë£Œ: {len(self.combined_data)}ê°œ ë¡œê·¸")

    def _convert_to_combined_no_comma(self):
        """ë¡œê·¸ ë°ì´í„°ë¥¼ ì‰¼í‘œ ì œê±°ëœ ê²°í•© í˜•íƒœë¡œ ë³€í™˜ (ê° value ë‚´ë¶€ì˜ ì‰¼í‘œë§Œ ì œê±°, value ê°„ êµ¬ë¶„ ì‰¼í‘œëŠ” ìœ ì§€)"""
        self.combined_data_no_comma = []
        
        # ì œì™¸í•  ê¸°ë³¸ í•„ë“œë“¤ (ë©”íƒ€ë°ì´í„°)
        exclude_fields = {'log_line', 'logtime', 'clientip', 'useragent', 'status', 
                        #  'page_id', 'click_type', 'act_type'
                         ## 6/13 ìˆ˜ì •ì •
                         } 
        
        for log_entry in self.input_data:
            log_no = log_entry.get('log_line', 0)
            
            # ê¸°ë³¸ ì •ë³´ ì¶”ì¶œ
            page_id = log_entry.get('page_id', '')
            click_type = log_entry.get('click_type', '')
            act_type = log_entry.get('act_type', '')
            click_text = log_entry.get('click_text', '')
            
            # LOGTIME í¬ë§· ë³€í™˜
            raw_logtime = log_entry.get('logtime', '')
            formatted_logtime = self._format_logtime(raw_logtime)
            
            # ë‚˜ë¨¸ì§€ íŒŒë¼ë¯¸í„°ë“¤ ìˆ˜ì§‘
            keys = []
            values = []
            
            for key, value in log_entry.items():
                if key not in exclude_fields:
                    keys.append(key)
                    # ê° ê°œë³„ value ë‚´ë¶€ì˜ ì‰¼í‘œë§Œ ì œê±° (CSV íŒŒì‹± ë¬¸ì œ í•´ê²°)
                    clean_value = str(value).replace(',', '') if value is not None else ''
                    values.append(clean_value)
            
            # í‚¤ì™€ ê°’ ëª¨ë‘ ì‰¼í‘œë¡œ êµ¬ë¶„ (í•˜ì§€ë§Œ ê° value ë‚´ë¶€ ì‰¼í‘œëŠ” ì´ë¯¸ ì œê±°ë¨)
            keys_combined = ', '.join(keys) if keys else ''
            values_combined = ', '.join(values) if values else ''  # CSV ë§¤ì¹­ì„ ìœ„í•´ ì‰¼í‘œ êµ¬ë¶„ ìœ ì§€
            
            combined_entry = {
                'no': log_no,
                # 'log_time': formatted_logtime,
                'page_id': page_id,
                'click_type': click_type,
                'act_type': act_type,
                'keys_combined': keys_combined,
                'values_combined': values_combined,
                'key_count': len(keys)
            }
            
            # click_textê°€ ìˆìœ¼ë©´ ì¶”ê°€ (ë‚´ë¶€ ì‰¼í‘œ ì œê±°)
            if click_text:
                combined_entry['click_text'] = str(click_text).replace(',', '')
            
            self.combined_data_no_comma.append(combined_entry)
        
        print(f"âœ… ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„° ë³€í™˜ ì™„ë£Œ: {len(self.combined_data_no_comma)}ê°œ ë¡œê·¸")
        print(f"   ğŸ“Œ ê° value ë‚´ë¶€ì˜ ì‰¼í‘œëŠ” ì œê±°, value ê°„ êµ¬ë¶„ ì‰¼í‘œëŠ” ìœ ì§€ (CSV ë§¤ì¹­ ìµœì í™”)")

    def _convert_to_combined_no_comma_deduplicated(self):
        """ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„°ì—ì„œ ì¤‘ë³µì„ ì œê±°í•œ í˜•íƒœë¡œ ë³€í™˜ (key_countê°€ í° ê²ƒ ìš°ì„ )"""
        if not self.combined_data_no_comma:
            print("âš ï¸ ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„°ê°€ ì—†ì–´ ì¤‘ë³µ ì œê±°ë¥¼ ìˆ˜í–‰í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        # pandas DataFrameìœ¼ë¡œ ë³€í™˜
        df = pd.DataFrame(self.combined_data_no_comma)
        
        # ì¤‘ë³µ ì œê±° ì „ ê°œìˆ˜
        original_count = len(df)
        
        # page_id, click_type, act_typeì„ ê¸°ì¤€ìœ¼ë¡œ ì¤‘ë³µ ì œê±°
        # click_textê°€ ìˆëŠ” ê²½ìš° ì¶”ê°€ë¡œ í¬í•¨
        dedup_columns = ['page_id', 'click_type', 'act_type']
        
        # click_text ì»¬ëŸ¼ì´ ì¡´ì¬í•˜ê³  ë¹„ì–´ìˆì§€ ì•Šì€ í–‰ì´ ìˆìœ¼ë©´ ì¶”ê°€
        if 'click_text' in df.columns and df['click_text'].notna().any() and (df['click_text'] != '').any():
            dedup_columns.append('click_text')
        
        # key_count ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬ (í° ê°’ì´ ë¨¼ì € ì˜¤ë„ë¡)
        df_sorted = df.sort_values(['key_count'] + dedup_columns, ascending=[False] + [True] * len(dedup_columns))
        
        # ì¤‘ë³µ ì œê±° (key_countê°€ í° ê²ƒì´ ë¨¼ì € ì˜¤ë¯€ë¡œ keep='first'ë¡œ í° ê°’ì´ ë‚¨ìŒ)
        df_deduplicated = df_sorted.drop_duplicates(subset=dedup_columns, keep='first')
        
        # ì›ë˜ ìˆœì„œëŒ€ë¡œ ì¬ì •ë ¬ (no ê¸°ì¤€)
        df_deduplicated = df_deduplicated.sort_values('no')
        
        # ì¤‘ë³µ ì œê±° í›„ ê°œìˆ˜
        deduplicated_count = len(df_deduplicated)
        removed_count = original_count - deduplicated_count
        
        # DataFrameì„ ë‹¤ì‹œ ë”•ì…”ë„ˆë¦¬ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
        self.combined_data_no_comma_deduplicated = df_deduplicated.to_dict('records')
        
        print(f"âœ… ì‰¼í‘œ ì œê±°ëœ ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±° ì™„ë£Œ (key_count í° ê²ƒ ìš°ì„ ):")
        print(f"   ì›ë³¸: {original_count}ê°œ")
        print(f"   ì¤‘ë³µ ì œê±° í›„: {deduplicated_count}ê°œ")
        print(f"   ì œê±°ëœ ì¤‘ë³µ: {removed_count}ê°œ")
        print(f"   ì¤‘ë³µ ì œê±° ê¸°ì¤€: {', '.join(dedup_columns)} + key_count í° ê²ƒ ìš°ì„ ")

    def _convert_to_combined_deduplicated(self):
        """ê²°í•© ë°ì´í„°ì—ì„œ ì¤‘ë³µì„ ì œê±°í•œ í˜•íƒœë¡œ ë³€í™˜ (key_countê°€ í° ê²ƒ ìš°ì„ )"""
        if not self.combined_data:
            print("âš ï¸ ê²°í•© ë°ì´í„°ê°€ ì—†ì–´ ì¤‘ë³µ ì œê±°ë¥¼ ìˆ˜í–‰í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        # pandas DataFrameìœ¼ë¡œ ë³€í™˜
        df = pd.DataFrame(self.combined_data)
        
        # ì¤‘ë³µ ì œê±° ì „ ê°œìˆ˜
        original_count = len(df)
        
        # page_id, click_type, act_typeì„ ê¸°ì¤€ìœ¼ë¡œ ì¤‘ë³µ ì œê±°
        # click_textê°€ ìˆëŠ” ê²½ìš° ì¶”ê°€ë¡œ í¬í•¨
        dedup_columns = ['page_id', 'click_type', 'act_type']
        
        # click_text ì»¬ëŸ¼ì´ ì¡´ì¬í•˜ê³  ë¹„ì–´ìˆì§€ ì•Šì€ í–‰ì´ ìˆìœ¼ë©´ ì¶”ê°€
        if 'click_text' in df.columns and df['click_text'].notna().any() and (df['click_text'] != '').any():
            dedup_columns.append('click_text')
        
        # key_count ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬ (í° ê°’ì´ ë¨¼ì € ì˜¤ë„ë¡)
        df_sorted = df.sort_values(['key_count'] + dedup_columns, ascending=[False] + [True] * len(dedup_columns))
        
        # ì¤‘ë³µ ì œê±° (key_countê°€ í° ê²ƒì´ ë¨¼ì € ì˜¤ë¯€ë¡œ keep='first'ë¡œ í° ê°’ì´ ë‚¨ìŒ)
        df_deduplicated = df_sorted.drop_duplicates(subset=dedup_columns, keep='first')
        
        # ì›ë˜ ìˆœì„œëŒ€ë¡œ ì¬ì •ë ¬ (no ê¸°ì¤€)
        df_deduplicated = df_deduplicated.sort_values('no')
        
        # ì¤‘ë³µ ì œê±° í›„ ê°œìˆ˜
        deduplicated_count = len(df_deduplicated)
        removed_count = original_count - deduplicated_count
        
        # DataFrameì„ ë‹¤ì‹œ ë”•ì…”ë„ˆë¦¬ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
        self.deduplicated_data = df_deduplicated.to_dict('records')
        
        print(f"âœ… ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±° ì™„ë£Œ (key_count í° ê²ƒ ìš°ì„ ):")
        print(f"   ì›ë³¸: {original_count}ê°œ")
        print(f"   ì¤‘ë³µ ì œê±° í›„: {deduplicated_count}ê°œ")
        print(f"   ì œê±°ëœ ì¤‘ë³µ: {removed_count}ê°œ")
        print(f"   ì¤‘ë³µ ì œê±° ê¸°ì¤€: {', '.join(dedup_columns)} + key_count í° ê²ƒ ìš°ì„ ")

    def _convert_to_detailed_deduplicated(self):
        """ìƒì„¸ ë°ì´í„°ì—ì„œ ì¤‘ë³µ ì œê±°í•œ í˜•íƒœë¡œ ë³€í™˜ (ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±°ëœ log_no ê¸°ì¤€ìœ¼ë¡œ í•„í„°ë§)"""
        if not self.deduplicated_data:
            print("âš ï¸ ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±° ë°ì´í„°ê°€ ì—†ì–´ ìƒì„¸ ì¤‘ë³µ ì œê±°ë¥¼ ìˆ˜í–‰í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        # ê²°í•© í˜•íƒœì—ì„œ ì¤‘ë³µ ì œê±°ëœ log_noë“¤ ì¶”ì¶œ
        deduplicated_log_nos = set(item['no'] for item in self.deduplicated_data)
        
        print(f"âœ… ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±°ì—ì„œ ìœ íš¨í•œ log_no: {len(deduplicated_log_nos)}ê°œ")
        
        self.detailed_deduplicated_data = []
        
        # ì œì™¸í•  ê¸°ë³¸ í•„ë“œë“¤ (ë©”íƒ€ë°ì´í„°) - _convert_to_detailedì™€ ë™ì¼
        exclude_fields = {'log_line', 'logtime', 'clientip', 'useragent', 'status'}
        
        for log_entry in self.input_data:
            log_no = log_entry.get('log_line', 0)
            
            # ê²°í•© í˜•íƒœì—ì„œ ì¤‘ë³µ ì œê±°ëœ log_noë§Œ ì²˜ë¦¬
            if log_no not in deduplicated_log_nos:
                continue
            
            # ê¸°ë³¸ ì •ë³´ ì¶”ì¶œ - _convert_to_detailedì™€ ë™ì¼
            page_id = log_entry.get('page_id', '')
            click_type = log_entry.get('click_type', '')
            act_type = log_entry.get('act_type', '')
            click_text = log_entry.get('click_text', '')
            
            # LOGTIME í¬ë§· ë³€í™˜ - _convert_to_detailedì™€ ë™ì¼
            raw_logtime = log_entry.get('logtime', '')
            formatted_logtime = self._format_logtime(raw_logtime)
            
            # ë‚˜ë¨¸ì§€ íŒŒë¼ë¯¸í„°ë“¤ì„ key-value ìŒìœ¼ë¡œ ì²˜ë¦¬ - _convert_to_detailedì™€ ë™ì¼
            for key, value in log_entry.items():
                if key in exclude_fields:
                    continue
                
                # ê¸°ë³¸ í•„ë“œë“¤ë„ key-value ìŒì— í¬í•¨ (ì‚¬ìš©ì ìš”ì²­)
                # page_id, click_type, act_typeë„ ì´ì œ key-valueë¡œ í‘œì‹œë¨
                
                self.detailed_deduplicated_data.append({
                    'no': log_no,
                    'log_time': formatted_logtime,
                    'page_id': page_id,
                    'click_type': click_type,
                    'act_type': act_type,
                    'click_text': click_text,
                    'key': key,
                    'value': str(value) if value is not None else ''
                })
        
        print(f"âœ… ìƒì„¸ í˜•íƒœ ì¤‘ë³µ ì œê±° ì™„ë£Œ (ê²°í•© í˜•íƒœ ê¸°ì¤€ í•„í„°ë§):")
        print(f"   í•„í„°ë§ëœ key-value ìŒ: {len(self.detailed_deduplicated_data)}ê°œ")
        print(f"   ê¸°ì¤€: ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±°ëœ {len(deduplicated_log_nos)}ê°œ ë¡œê·¸")

    def _extract_click_text_from_keys(self, row):
        """keys_combinedì—ì„œ click_text ê°’ì„ ì¶”ì¶œí•˜ë ¤ê³  ì‹œë„"""
        try:
            keys = row.get('keys_combined', '').split(', ')
            values = row.get('values_combined', '').split(', ')
            
            if len(keys) == len(values):
                for i, key in enumerate(keys):
                    if key.strip() == 'click_text':
                        return values[i].strip() if i < len(values) else ''
            return ''
        except:
            return ''

    def _save_detailed_excel(self, output_path: str):
        """ìƒì„¸ ë°ì´í„°ë¥¼ Excel íŒŒì¼ë¡œ ì €ì¥"""
        try:
            df = pd.DataFrame(self.detailed_data)
            
            if df.empty:
                print("âš ï¸ ì €ì¥í•  ìƒì„¸ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
                return
            
            # ì»¬ëŸ¼ ìˆœì„œ ì •ì˜ (log_timeì„ ë‘ ë²ˆì§¸ ì—´ë¡œ)
            column_order = ['no', 'log_time', 'page_id', 'act_type', 'click_type', 'click_text', 'key', 'value']
            
            # ì¡´ì¬í•˜ëŠ” ì»¬ëŸ¼ë§Œ ì„ íƒ
            available_columns = [col for col in column_order if col in df.columns]
            df = df[available_columns]
            
            # Excel íŒŒì¼ ì €ì¥
            df.to_excel(output_path, index=False, engine='openpyxl')
            
            print(f"âœ… ìƒì„¸ Excel íŒŒì¼ ì €ì¥: {output_path}")
            print(f"   ì´ {len(self.detailed_data)}ê°œ key-value ìŒ")
            
        except Exception as e:
            print(f"âŒ íŒŒì¼ ì €ì¥ ì˜¤ë¥˜: {e}")
            raise

    def _save_detailed_deduplicated_excel(self, output_path: str):
        """ìƒì„¸ ì¤‘ë³µ ì œê±°ëœ ë°ì´í„°ë¥¼ Excel íŒŒì¼ë¡œ ì €ì¥"""
        try:
            df = pd.DataFrame(self.detailed_deduplicated_data)
            
            if df.empty:
                print("âš ï¸ ì €ì¥í•  ìƒì„¸ ì¤‘ë³µ ì œê±° ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
                return
            
            # ì»¬ëŸ¼ ìˆœì„œ ì •ì˜ (log_timeì„ ë‘ ë²ˆì§¸ ì—´ë¡œ)
            column_order = ['no', 'log_time', 'page_id', 'click_type', 'act_type', 'click_text', 'key', 'value']
            
            # ì¡´ì¬í•˜ëŠ” ì»¬ëŸ¼ë§Œ ì„ íƒ
            available_columns = [col for col in column_order if col in df.columns]
            df = df[available_columns]
            
            # Excel íŒŒì¼ ì €ì¥
            df.to_excel(output_path, index=False, engine='openpyxl')
            
            print(f"âœ… ìƒì„¸ ì¤‘ë³µ ì œê±° Excel íŒŒì¼ ì €ì¥: {output_path}")
            print(f"   ì´ {len(self.detailed_deduplicated_data)}ê°œ key-value ìŒ (ì¤‘ë³µ ì œê±°ë¨)")
            
        except Exception as e:
            print(f"âŒ íŒŒì¼ ì €ì¥ ì˜¤ë¥˜: {e}")
            raise

    def _save_combined_deduplicated_excel(self, output_path: str):
        """ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±°ëœ ë°ì´í„°ë¥¼ Excel íŒŒì¼ë¡œ ì €ì¥"""
        try:
            df = pd.DataFrame(self.deduplicated_data)
            
            if df.empty:
                print("âš ï¸ ì €ì¥í•  ì¤‘ë³µ ì œê±° ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
                return
            
            # ì»¬ëŸ¼ ìˆœì„œ ì •ì˜ (log_timeì„ ë‘ ë²ˆì§¸ ì—´ë¡œ)
            column_order = ['no', 'log_time', 'page_id', 'click_type', 'act_type', 'keys_combined', 'values_combined', 'key_count']
            
            # click_text ì»¬ëŸ¼ì´ ìˆìœ¼ë©´ ì¶”ê°€
            if 'click_text' in df.columns:
                column_order.insert(5, 'click_text')
            
            # ì¡´ì¬í•˜ëŠ” ì»¬ëŸ¼ë§Œ ì„ íƒ
            available_columns = [col for col in column_order if col in df.columns]
            df = df[available_columns]
            
            # Excel íŒŒì¼ ì €ì¥
            df.to_excel(output_path, index=False, engine='openpyxl')
            
            print(f"âœ… ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±° Excel íŒŒì¼ ì €ì¥: {output_path}")
            print(f"   ì´ {len(self.deduplicated_data)}ê°œ ë¡œê·¸ (ì¤‘ë³µ ì œê±°ë¨)")
            
        except Exception as e:
            print(f"âŒ íŒŒì¼ ì €ì¥ ì˜¤ë¥˜: {e}")
            raise

    def _save_combined_excel(self, output_path: str):
        """ê²°í•© ë°ì´í„°ë¥¼ Excel íŒŒì¼ë¡œ ì €ì¥"""
        try:
            df = pd.DataFrame(self.combined_data)
            
            if df.empty:
                print("âš ï¸ ì €ì¥í•  ê²°í•© ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
                return
            
            # ì»¬ëŸ¼ ìˆœì„œ ì •ì˜ (log_timeì„ ë‘ ë²ˆì§¸ ì—´ë¡œ)
            column_order = ['no', 'log_time', 'page_id', 'click_type', 'act_type', 'keys_combined', 'values_combined', 'key_count']
            
            # click_text ì»¬ëŸ¼ì´ ìˆìœ¼ë©´ ì¶”ê°€
            if 'click_text' in df.columns:
                column_order.insert(5, 'click_text')
            
            # ì¡´ì¬í•˜ëŠ” ì»¬ëŸ¼ë§Œ ì„ íƒ
            available_columns = [col for col in column_order if col in df.columns]
            df = df[available_columns]
            
            # Excel íŒŒì¼ ì €ì¥
            df.to_excel(output_path, index=False, engine='openpyxl')
            
            print(f"âœ… ê²°í•© Excel íŒŒì¼ ì €ì¥: {output_path}")
            print(f"   ì´ {len(self.combined_data)}ê°œ ë¡œê·¸")
            
        except Exception as e:
            print(f"âŒ íŒŒì¼ ì €ì¥ ì˜¤ë¥˜: {e}")
            raise

    def _save_combined_no_comma_excel(self, output_path: str):
        """value ë‚´ë¶€ ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„°ë¥¼ Excel íŒŒì¼ë¡œ ì €ì¥"""
        try:
            df = pd.DataFrame(self.combined_data_no_comma)
            
            if df.empty:
                print("âš ï¸ ì €ì¥í•  value ë‚´ë¶€ ì‰¼í‘œ ì œê±° ê²°í•© ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
                return
            
            # ì»¬ëŸ¼ ìˆœì„œ ì •ì˜ (log_timeì„ ë‘ ë²ˆì§¸ ì—´ë¡œ)
            column_order = ['no', 'log_time', 'page_id', 'click_type', 'act_type', 'keys_combined', 'values_combined', 'key_count']
            
            # click_text ì»¬ëŸ¼ì´ ìˆìœ¼ë©´ ì¶”ê°€
            if 'click_text' in df.columns:
                column_order.insert(5, 'click_text')
            
            # ì¡´ì¬í•˜ëŠ” ì»¬ëŸ¼ë§Œ ì„ íƒ
            available_columns = [col for col in column_order if col in df.columns]
            df = df[available_columns]
            
            # Excel íŒŒì¼ ì €ì¥
            df.to_excel(output_path, index=False, engine='openpyxl')
            
            print(f"âœ… CSV ë§¤ì¹­ ìµœì í™”ëœ Excel íŒŒì¼ ì €ì¥: {output_path}")
            print(f"   ì´ {len(self.combined_data_no_comma)}ê°œ ë¡œê·¸ (ê° value ë‚´ë¶€ ì‰¼í‘œ ì œê±°, value ê°„ êµ¬ë¶„ ì‰¼í‘œ ìœ ì§€)")
            
        except Exception as e:
            print(f"âŒ íŒŒì¼ ì €ì¥ ì˜¤ë¥˜: {e}")
            raise

    def _save_combined_no_comma_deduplicated_excel(self, output_path: str):
        """ì‰¼í‘œ ì œê±°ëœ ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±°ëœ ë°ì´í„°ë¥¼ Excel íŒŒì¼ë¡œ ì €ì¥"""
        try:
            df = pd.DataFrame(self.combined_data_no_comma_deduplicated)
            
            if df.empty:
                print("âš ï¸ ì €ì¥í•  ì‰¼í‘œ ì œê±°ëœ ì¤‘ë³µ ì œê±° ê²°í•© ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
                return
            
            # ì»¬ëŸ¼ ìˆœì„œ ì •ì˜ (log_timeì„ ë‘ ë²ˆì§¸ ì—´ë¡œ)
            column_order = ['no', 'page_id', 'click_type', 'act_type', 'keys_combined', 'values_combined', 'key_count']
            
            # click_text ì»¬ëŸ¼ì´ ìˆìœ¼ë©´ ì¶”ê°€
            if 'click_text' in df.columns:
                column_order.insert(4, 'click_text')
            
            # ì¡´ì¬í•˜ëŠ” ì»¬ëŸ¼ë§Œ ì„ íƒ
            available_columns = [col for col in column_order if col in df.columns]
            df = df[available_columns]
            
            # Excel íŒŒì¼ ì €ì¥
            df.to_excel(output_path, index=False, engine='openpyxl')
            
            print(f"âœ… ì‰¼í‘œ ì œê±°ëœ ì¤‘ë³µ ì œê±° Excel íŒŒì¼ ì €ì¥: {output_path}")
            print(f"   ì´ {len(self.combined_data_no_comma_deduplicated)}ê°œ ë¡œê·¸ (ê° value ë‚´ë¶€ ì‰¼í‘œ ì œê±° + ì¤‘ë³µ ì œê±°)")
            
        except Exception as e:
            print(f"âŒ íŒŒì¼ ì €ì¥ ì˜¤ë¥˜: {e}")
            raise

    def get_conversion_summary(self) -> Dict:
        """ë³€í™˜ ê²°ê³¼ ìš”ì•½ ì •ë³´ ë°˜í™˜"""
        if not self.detailed_data:
            return {"error": "ë³€í™˜ëœ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤."}
        
        # í†µê³„ ê³„ì‚°
        total_logs = len(self.input_data)
        total_key_values = len(self.detailed_data)
        avg_keys_per_log = total_key_values / total_logs if total_logs > 0 else 0
        
        # ê°€ì¥ ë§ì€ í‚¤ë¥¼ ê°€ì§„ ë¡œê·¸ ì°¾ê¸°
        key_counts = {}
        for item in self.detailed_data:
            log_no = item['no']
            key_counts[log_no] = key_counts.get(log_no, 0) + 1
        
        max_keys = max(key_counts.values()) if key_counts else 0
        min_keys = min(key_counts.values()) if key_counts else 0
        
        # ê³ ìœ í•œ í‚¤ ìˆ˜ ê³„ì‚°
        unique_keys = set(item['key'] for item in self.detailed_data)
        
        # ì¤‘ë³µ ì œê±° í†µê³„
        combined_count = len(self.combined_data) if self.combined_data else 0
        combined_deduplicated_count = len(self.deduplicated_data) if self.deduplicated_data else 0
        combined_duplicates_removed = combined_count - combined_deduplicated_count
        
        detailed_count = len(self.detailed_data) if self.detailed_data else 0
        detailed_deduplicated_count = len(self.detailed_deduplicated_data) if self.detailed_deduplicated_data else 0
        detailed_duplicates_removed = detailed_count - detailed_deduplicated_count
        
        # ì‰¼í‘œ ì œê±° í†µê³„ ì¶”ê°€
        combined_no_comma_count = len(self.combined_data_no_comma) if self.combined_data_no_comma else 0
        combined_no_comma_deduplicated_count = len(self.combined_data_no_comma_deduplicated) if self.combined_data_no_comma_deduplicated else 0
        combined_no_comma_duplicates_removed = combined_no_comma_count - combined_no_comma_deduplicated_count
        
        return {
            "total_logs": total_logs,
            "total_key_value_pairs": total_key_values,
            "average_keys_per_log": round(avg_keys_per_log, 2),
            "max_keys_in_single_log": max_keys,
            "min_keys_in_single_log": min_keys,
            "unique_keys_count": len(unique_keys),
            "unique_keys": sorted(list(unique_keys)),
            "combined_logs_count": combined_count,
            "combined_deduplicated_count": combined_deduplicated_count,
            "combined_duplicates_removed": combined_duplicates_removed,
            "combined_no_comma_count": combined_no_comma_count,
            "combined_no_comma_deduplicated_count": combined_no_comma_deduplicated_count,
            "combined_no_comma_duplicates_removed": combined_no_comma_duplicates_removed,
            "detailed_key_value_pairs": detailed_count,
            "detailed_deduplicated_count": detailed_deduplicated_count,
            "detailed_duplicates_removed": detailed_duplicates_removed
        }

    def preview_data(self, num_rows: int = 10):
        """ë³€í™˜ëœ ìƒì„¸ ë°ì´í„° ë¯¸ë¦¬ë³´ê¸°"""
        print(f"\nğŸ“‹ ìƒì„¸ ë°ì´í„° ë¯¸ë¦¬ë³´ê¸° (ì²˜ìŒ {num_rows}ê°œ í–‰):")
        print("-" * 160)
        print(f"{'No':3} | {'Log Time':19} | {'Page ID':20} | {'Click Type':12} | {'Act Type':12} | {'Click Text':15} | {'Key':15} | {'Value':20}")
        print("-" * 160)
        
        if self.detailed_data:
            for item in self.detailed_data[:num_rows]:
                log_time = item.get('log_time', '')[:19] if len(item.get('log_time', '')) > 19 else item.get('log_time', '')
                page_id = item['page_id'][:20] if len(item['page_id']) > 20 else item['page_id']
                click_type = item['click_type'][:12] if len(item['click_type']) > 12 else item['click_type']
                act_type = item['act_type'][:12] if len(item['act_type']) > 12 else item['act_type']
                click_text = str(item.get('click_text', ''))[:15] if len(str(item.get('click_text', ''))) > 15 else str(item.get('click_text', ''))
                key = item['key'][:15] if len(item['key']) > 15 else item['key']
                value = item['value'][:20] if len(item['value']) > 20 else item['value']
                
                print(f"{item['no']:3} | {log_time:19} | {page_id:20} | {click_type:12} | {act_type:12} | {click_text:15} | {key:15} | {value:20}")
            
            if len(self.detailed_data) > num_rows:
                print(f"... ì™¸ {len(self.detailed_data) - num_rows}ê°œ í–‰")
        else:
            print("ë³€í™˜ëœ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")

    def preview_combined_no_comma_deduplicated_data(self, num_rows: int = 5):
        """ì‰¼í‘œ ì œê±°ëœ ì¤‘ë³µ ì œê±° ê²°í•© ë°ì´í„° ë¯¸ë¦¬ë³´ê¸°"""
        print(f"\nğŸ“‹ value ë‚´ë¶€ ì‰¼í‘œ ì œê±° + ì¤‘ë³µ ì œê±°ëœ ê²°í•© ë°ì´í„° ë¯¸ë¦¬ë³´ê¸° (ì²˜ìŒ {num_rows}ê°œ í–‰):")
        print(f"   ğŸ“Œ CSV ë§¤ì¹­ìš©: ê° value ë‚´ë¶€ ì‰¼í‘œ ì œê±° + ì¤‘ë³µ ì œê±°, value ê°„ êµ¬ë¶„ ì‰¼í‘œ ìœ ì§€")
        print("-" * 170)
        
        # í—¤ë” ë™ì  ìƒì„±
        has_click_text = self.combined_data_no_comma_deduplicated and any('click_text' in item for item in self.combined_data_no_comma_deduplicated[:3])
        if has_click_text:
            print(f"{'No':3} | {'Log Time':19} | {'Page ID':15} | {'Click Type':12} | {'Act Type':12} | {'Click Text':12} | {'Keys':20} | {'Values (ë‚´ë¶€ì‰¼í‘œì œê±°)':20} | {'Count':5}")
        else:
            print(f"{'No':3} | {'Log Time':19} | {'Page ID':20} | {'Click Type':12} | {'Act Type':12} | {'Keys':25} | {'Values (ë‚´ë¶€ì‰¼í‘œì œê±°)':25} | {'Count':5}")
        
        print("-" * 170)
        
        if self.combined_data_no_comma_deduplicated:
            for item in self.combined_data_no_comma_deduplicated[:num_rows]:
                log_time = item.get('log_time', '')[:19] if len(item.get('log_time', '')) > 19 else item.get('log_time', '')
                page_id = item['page_id'][:15 if has_click_text else 20] if len(item['page_id']) > (15 if has_click_text else 20) else item['page_id']
                click_type = item['click_type'][:12] if len(item['click_type']) > 12 else item['click_type']
                act_type = item['act_type'][:12] if len(item['act_type']) > 12 else item['act_type']
                
                if 'click_text' in item and item['click_text']:
                    click_text = item['click_text'][:12] if len(str(item['click_text'])) > 12 else str(item['click_text'])
                    keys = item['keys_combined'][:20] if len(item['keys_combined']) > 20 else item['keys_combined']
                    values = item['values_combined'][:20] if len(item['values_combined']) > 20 else item['values_combined']
                    print(f"{item['no']:3} | {log_time:19} | {page_id:15} | {click_type:12} | {act_type:12} | {click_text:12} | {keys:20} | {values:20} | {item['key_count']:5}")
                else:
                    keys = item['keys_combined'][:25] if len(item['keys_combined']) > 25 else item['keys_combined']
                    values = item['values_combined'][:25] if len(item['values_combined']) > 25 else item['values_combined']
                    print(f"{item['no']:3} | {log_time:19} | {page_id:20} | {click_type:12} | {act_type:12} | {keys:25} | {values:25} | {item['key_count']:5}")
            
            if len(self.combined_data_no_comma_deduplicated) > num_rows:
                print(f"... ì™¸ {len(self.combined_data_no_comma_deduplicated) - num_rows}ê°œ í–‰")
        else:
            print("ë³€í™˜ëœ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")

    def preview_combined_no_comma_data(self, num_rows: int = 5):
        """ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„° ë¯¸ë¦¬ë³´ê¸°"""
        print(f"\nğŸ“‹ value ë‚´ë¶€ ì‰¼í‘œ ì œê±°ëœ ê²°í•© ë°ì´í„° ë¯¸ë¦¬ë³´ê¸° (ì²˜ìŒ {num_rows}ê°œ í–‰):")
        print(f"   ğŸ“Œ CSV ë§¤ì¹­ìš©: ê° value ë‚´ë¶€ ì‰¼í‘œ ì œê±°, value ê°„ êµ¬ë¶„ ì‰¼í‘œ ìœ ì§€")
        print("-" * 170)
        print(f"{'No':3} | {'Log Time':19} | {'Page ID':20} | {'Click Type':12} | {'Act Type':12} | {'Keys':30} | {'Values (ë‚´ë¶€ì‰¼í‘œì œê±°)':30} | {'Count':5}")
        print("-" * 170)
        
        if self.combined_data_no_comma:
            for item in self.combined_data_no_comma[:num_rows]:
                log_time = item.get('log_time', '')[:19] if len(item.get('log_time', '')) > 19 else item.get('log_time', '')
                page_id = item['page_id'][:20] if len(item['page_id']) > 20 else item['page_id']
                click_type = item['click_type'][:12] if len(item['click_type']) > 12 else item['click_type']
                act_type = item['act_type'][:12] if len(item['act_type']) > 12 else item['act_type']
                keys = item['keys_combined'][:30] if len(item['keys_combined']) > 30 else item['keys_combined']
                values = item['values_combined'][:30] if len(item['values_combined']) > 30 else item['values_combined']
                
                print(f"{item['no']:3} | {log_time:19} | {page_id:20} | {click_type:12} | {act_type:12} | {keys:30} | {values:30} | {item['key_count']:5}")
            
            if len(self.combined_data_no_comma) > num_rows:
                print(f"... ì™¸ {len(self.combined_data_no_comma) - num_rows}ê°œ í–‰")
        else:
            print("ë³€í™˜ëœ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")

    def preview_combined_deduplicated_data(self, num_rows: int = 5):
        """ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±°ëœ ë°ì´í„° ë¯¸ë¦¬ë³´ê¸°"""
        print(f"\nğŸ“‹ ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±°ëœ ë°ì´í„° ë¯¸ë¦¬ë³´ê¸° (ì²˜ìŒ {num_rows}ê°œ í–‰):")
        print("-" * 170)
        
        # í—¤ë” ë™ì  ìƒì„±
        has_click_text = self.deduplicated_data and any('click_text' in item for item in self.deduplicated_data[:3])
        if has_click_text:
            print(f"{'No':3} | {'Log Time':19} | {'Page ID':15} | {'Click Type':12} | {'Act Type':12} | {'Click Text':12} | {'Keys':20} | {'Values':20} | {'Count':5}")
        else:
            print(f"{'No':3} | {'Log Time':19} | {'Page ID':20} | {'Click Type':12} | {'Act Type':12} | {'Keys':25} | {'Values':25} | {'Count':5}")
        
        print("-" * 170)
        
        if self.deduplicated_data:
            for item in self.deduplicated_data[:num_rows]:
                log_time = item.get('log_time', '')[:19] if len(item.get('log_time', '')) > 19 else item.get('log_time', '')
                page_id = item['page_id'][:15 if has_click_text else 20] if len(item['page_id']) > (15 if has_click_text else 20) else item['page_id']
                click_type = item['click_type'][:12] if len(item['click_type']) > 12 else item['click_type']
                act_type = item['act_type'][:12] if len(item['act_type']) > 12 else item['act_type']
                
                if 'click_text' in item and item['click_text']:
                    click_text = item['click_text'][:12] if len(str(item['click_text'])) > 12 else str(item['click_text'])
                    keys = item['keys_combined'][:20] if len(item['keys_combined']) > 20 else item['keys_combined']
                    values = item['values_combined'][:20] if len(item['values_combined']) > 20 else item['values_combined']
                    print(f"{item['no']:3} | {log_time:19} | {page_id:15} | {click_type:12} | {act_type:12} | {click_text:12} | {keys:20} | {values:20} | {item['key_count']:5}")
                else:
                    keys = item['keys_combined'][:25] if len(item['keys_combined']) > 25 else item['keys_combined']
                    values = item['values_combined'][:25] if len(item['values_combined']) > 25 else item['values_combined']
                    print(f"{item['no']:3} | {log_time:19} | {page_id:20} | {click_type:12} | {act_type:12} | {keys:25} | {values:25} | {item['key_count']:5}")
            
            if len(self.deduplicated_data) > num_rows:
                print(f"... ì™¸ {len(self.deduplicated_data) - num_rows}ê°œ í–‰")
        else:
            print("ë³€í™˜ëœ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")

    def preview_detailed_deduplicated_data(self, num_rows: int = 10):
        """ìƒì„¸ ì¤‘ë³µ ì œê±°ëœ ë°ì´í„° ë¯¸ë¦¬ë³´ê¸°"""
        print(f"\nğŸ“‹ ìƒì„¸ ì¤‘ë³µ ì œê±°ëœ ë°ì´í„° ë¯¸ë¦¬ë³´ê¸° (ì²˜ìŒ {num_rows}ê°œ í–‰):")
        print("-" * 160)
        print(f"{'No':3} | {'Log Time':19} | {'Page ID':20} | {'Click Type':12} | {'Act Type':12} | {'Click Text':15} | {'Key':15} | {'Value':20}")
        print("-" * 160)
        
        if self.detailed_deduplicated_data:
            for item in self.detailed_deduplicated_data[:num_rows]:
                log_time = item.get('log_time', '')[:19] if len(item.get('log_time', '')) > 19 else item.get('log_time', '')
                page_id = item['page_id'][:20] if len(item['page_id']) > 20 else item['page_id']
                click_type = item['click_type'][:12] if len(item['click_type']) > 12 else item['click_type']
                act_type = item['act_type'][:12] if len(item['act_type']) > 12 else item['act_type']
                click_text = str(item.get('click_text', ''))[:15] if len(str(item.get('click_text', ''))) > 15 else str(item.get('click_text', ''))
                key = item['key'][:15] if len(item['key']) > 15 else item['key']
                value = item['value'][:20] if len(item['value']) > 20 else item['value']
                
                print(f"{item['no']:3} | {log_time:19} | {page_id:20} | {click_type:12} | {act_type:12} | {click_text:15} | {key:15} | {value:20}")
            
            if len(self.detailed_deduplicated_data) > num_rows:
                print(f"... ì™¸ {len(self.detailed_deduplicated_data) - num_rows}ê°œ í–‰")
        else:
            print("ë³€í™˜ëœ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")

    def preview_combined_data(self, num_rows: int = 5):
        """ê²°í•© ë°ì´í„° ë¯¸ë¦¬ë³´ê¸°"""
        print(f"\nğŸ“‹ ê²°í•© ë°ì´í„° ë¯¸ë¦¬ë³´ê¸° (ì²˜ìŒ {num_rows}ê°œ í–‰):")
        print("-" * 170)
        print(f"{'No':3} | {'Log Time':19} | {'Page ID':20} | {'Click Type':12} | {'Act Type':12} | {'Keys':30} | {'Values':30} | {'Count':5}")
        print("-" * 170)
        
        if self.combined_data:
            for item in self.combined_data[:num_rows]:
                log_time = item.get('log_time', '')[:19] if len(item.get('log_time', '')) > 19 else item.get('log_time', '')
                page_id = item['page_id'][:20] if len(item['page_id']) > 20 else item['page_id']
                click_type = item['click_type'][:12] if len(item['click_type']) > 12 else item['click_type']
                act_type = item['act_type'][:12] if len(item['act_type']) > 12 else item['act_type']
                keys = item['keys_combined'][:30] if len(item['keys_combined']) > 30 else item['keys_combined']
                values = item['values_combined'][:30] if len(item['values_combined']) > 30 else item['values_combined']
                
                print(f"{item['no']:3} | {log_time:19} | {page_id:20} | {click_type:12} | {act_type:12} | {keys:30} | {values:30} | {item['key_count']:5}")
            
            if len(self.combined_data) > num_rows:
                print(f"... ì™¸ {len(self.combined_data) - num_rows}ê°œ í–‰")
        else:
            print("ë³€í™˜ëœ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")

    def get_unique_keys_analysis(self):
        """ê³ ìœ  í‚¤ ë¶„ì„"""
        if not self.detailed_data:
            print("ë¶„ì„í•  ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
            return
        
        # í‚¤ë³„ ë¹ˆë„ìˆ˜ ê³„ì‚°
        key_frequency = {}
        for item in self.detailed_data:
            key = item['key']
            key_frequency[key] = key_frequency.get(key, 0) + 1
        
        # ë¹ˆë„ìˆ˜ ìˆœìœ¼ë¡œ ì •ë ¬
        sorted_keys = sorted(key_frequency.items(), key=lambda x: x[1], reverse=True)
        
        print(f"\nğŸ“Š í‚¤ ë¶„ì„ (ì´ {len(sorted_keys)}ê°œ ê³ ìœ  í‚¤):")
        print("-" * 50)
        print(f"{'Key':30} | {'ë¹ˆë„ìˆ˜':10}")
        print("-" * 50)
        
        for key, freq in sorted_keys[:20]:  # ìƒìœ„ 20ê°œë§Œ í‘œì‹œ
            key_display = key[:30] if len(key) > 30 else key
            print(f"{key_display:30} | {freq:10}")
        
        if len(sorted_keys) > 20:
            print(f"... ì™¸ {len(sorted_keys) - 20}ê°œ í‚¤")


def run_log_converter_no_comma_example_with_fill_missing():
    """value ë‚´ë¶€ ì‰¼í‘œ ì œê±°ëœ ë¡œê·¸ ë³€í™˜ê¸° ì‚¬ìš© ì˜ˆì‹œ (ëˆ„ë½ëœ í–‰ ì±„ìš°ê¸° í¬í•¨)"""
    converter = LogToDetailedConverter()

    print("=== JSON ë¡œê·¸ë¥¼ CSV ë§¤ì¹­ ìµœì í™”ëœ Excel í˜•íƒœë¡œ ë³€í™˜ (ëˆ„ë½ëœ í–‰ ì±„ìš°ê¸° í¬í•¨) ===")
    print("ğŸ“Œ ëª©ì : ê° value ë‚´ë¶€ì˜ ì‰¼í‘œë§Œ ì œê±°í•˜ì—¬ CSV íŒŒì‹± ì‹œ key-value ë§¤ì¹­ ì •í™•ë„ í–¥ìƒ")
    print("ğŸ“Œ ë°©ë²•: value ê°„ êµ¬ë¶„ ì‰¼í‘œëŠ” ìœ ì§€, ê°œë³„ value ë‚´ë¶€ ì‰¼í‘œë§Œ ì œê±°")
    print("ğŸ“Œ ì¶”ê°€: ì¤‘ë³µ ì œê±° ë²„ì „ë„ í•¨ê»˜ ìƒì„±")
    print("ğŸ“Œ ìƒˆê¸°ëŠ¥: ëˆ„ë½ëœ no ë²ˆí˜¸ì— ë¹ˆ í–‰ ì¶”ê°€\n")
    
    # 1. ìƒì„¸ í˜•íƒœì™€ ì‰¼í‘œ ì œê±°ëœ ê²°í•© í˜•íƒœë¡œ ë³€í™˜
    converter.convert_log_to_combined_no_comma(
        "./log/log_file.txt",
        "./result/log_analysis_detailed.xlsx",
        "./result/log_analysis_combined_no_comma.xlsx",
        "./result/log_analysis_combined_no_comma_deduplicated.xlsx"
    )
    
    # 2. ëˆ„ë½ëœ ë²ˆí˜¸ ë¶„ì„
    converter.preview_missing_analysis()
    
    # 3. ëˆ„ë½ëœ í–‰ ì±„ìš°ê¸°
    converter.fill_all_missing_rows()
    
    # 4. ì±„ìš°ê¸° ì™„ë£Œëœ ë°ì´í„°ë¡œ ë‹¤ì‹œ ì €ì¥
    converter._save_detailed_excel("./result/log_analysis_detailed_filled.xlsx")
    converter._save_combined_no_comma_excel("./result/log_analysis_combined_no_comma_filled.xlsx")
    
    if converter.combined_data_no_comma_deduplicated:
        converter._save_combined_no_comma_deduplicated_excel("./result/log_analysis_combined_no_comma_deduplicated_filled.xlsx")
    
    # 5. ê²°ê³¼ ìš”ì•½ ì¶œë ¥
    summary = converter.get_conversion_summary()
    print(f"\nğŸ“Š ë³€í™˜ ê²°ê³¼ ìš”ì•½:")
    for key, value in summary.items():
        if key != 'unique_keys':
            print(f"   {key}: {value}")
    
    # 6. ë°ì´í„° ë¯¸ë¦¬ë³´ê¸°
    converter.preview_data(10)
    converter.preview_combined_no_comma_data(5)
    
    print(f"\nâœ… ëˆ„ë½ëœ í–‰ì´ ì±„ì›Œì§„ Excel íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤:")
    print(f"   - ìƒì„¸ í˜•íƒœ (ì±„ì›Œì§): ./result/log_analysis_detailed_filled.xlsx")
    print(f"   - CSV ë§¤ì¹­ ìµœì í™” (ì±„ì›Œì§): ./result/log_analysis_combined_no_comma_filled.xlsx")
    if converter.combined_data_no_comma_deduplicated:
        print(f"   - CSV ë§¤ì¹­ ìµœì í™” + ì¤‘ë³µ ì œê±° (ì±„ì›Œì§): ./result/log_analysis_combined_no_comma_deduplicated_filled.xlsx")


def run_log_converter_example():
    """ë¡œê·¸ ë³€í™˜ê¸° ì‚¬ìš© ì˜ˆì‹œ (ê¸°ì¡´ í•¨ìˆ˜)"""
    converter = LogToDetailedConverter()

    print("=== JSON ë¡œê·¸ë¥¼ Excel í˜•íƒœë¡œ ë³€í™˜ (JSON ì˜¤ë¥˜ ìë™ ìˆ˜ì • + ì¤‘ë³µ ì œê±° ê¸°ëŠ¥ í¬í•¨) ===\n")
    
    # ë¡œê·¸ íŒŒì¼ì„ ë„¤ ê°€ì§€ í˜•íƒœë¡œ ë³€í™˜
    converter.convert_log_to_all_formats(
        "./log/log_file.txt",                              # ì…ë ¥ JSON ë¡œê·¸ íŒŒì¼
        "./result/log_analysis_detailed.xlsx",                # ìƒì„¸ í˜•íƒœ ì¶œë ¥
        "./result/log_analysis_detailed_deduplicated.xlsx",   # ìƒì„¸ í˜•íƒœ ì¤‘ë³µ ì œê±° ì¶œë ¥
        "./result/log_analysis_combined.xlsx",                # ê²°í•© í˜•íƒœ ì¶œë ¥
        "./result/log_analysis_combined_deduplicated.xlsx"    # ê²°í•© í˜•íƒœ ì¤‘ë³µ ì œê±° ì¶œë ¥
    )
    
    # ê²°ê³¼ ìš”ì•½ ì¶œë ¥
    summary = converter.get_conversion_summary()
    print(f"\nğŸ“Š ë³€í™˜ ê²°ê³¼ ìš”ì•½:")
    for key, value in summary.items():
        if key != 'unique_keys':  # unique_keysëŠ” ë„ˆë¬´ ê¸¸ì–´ì„œ ë³„ë„ ì¶œë ¥
            print(f"   {key}: {value}")
    
    # ë°ì´í„° ë¯¸ë¦¬ë³´ê¸°
    converter.preview_data(10)
    converter.preview_detailed_deduplicated_data(10)
    converter.preview_combined_data(5)
    converter.preview_combined_deduplicated_data(5)
    
    # í‚¤ ë¶„ì„
    converter.get_unique_keys_analysis()


if __name__ == "__main__":
    # ì‰¼í‘œ ì œê±°ëœ ë²„ì „ ì‹¤í–‰
    run_log_converter_no_comma_example_with_fill_missing()
    
    print("\n" + "="*80 + "\n")
    
    # ê¸°ì¡´ ë²„ì „ë„ ì‹¤í–‰ (ì„ íƒì‚¬í•­)
    run_log_converter_example()
```
